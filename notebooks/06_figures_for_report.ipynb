# 06 — Figures for Report
Reproduce all 4 final figures in publication quality, save to outputs/figures/.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from config import INTERIM_DIR, PROCESSED_DIR, FIGURES_DIR
from plots import (
    plot_match_timeline,
    plot_momentum_bins,
    plot_real_vs_sim,
    plot_bootstrap_intervals,
)

os.makedirs(FIGURES_DIR, exist_ok=True)
print(f'Figures will be saved to: {FIGURES_DIR}')
```

```python
# Load all data
shots_df          = pd.read_csv(os.path.join(INTERIM_DIR, 'shots.csv'))
team_windows_df   = pd.read_csv(os.path.join(INTERIM_DIR, 'team_windows.csv'))
sim_windows_df    = pd.read_csv(os.path.join(INTERIM_DIR, 'simulated_team_windows.csv'))
bootstrap_df      = pd.read_json(os.path.join(PROCESSED_DIR, 'bootstrap_results.json'))

print(f'shots:       {len(shots_df):>6} rows')
print(f'real windows:{len(team_windows_df):>6} rows')
print(f'sim windows: {len(sim_windows_df):>6} rows')
print(f'bootstrap:   {len(bootstrap_df):>6} rows')
```

```python
# Fig 1: Match timeline
# Pick a match with interesting action
match_shot_counts = shots_df.groupby('match_id').size()
featured_match = match_shot_counts.idxmax()
print(f'Fig 1: timeline for match {featured_match} ({match_shot_counts[featured_match]} shots)')
plot_match_timeline(shots_df, featured_match,
                    save_path=os.path.join(FIGURES_DIR, 'fig1_match_timeline.png'))
```

```python
# Fig 2 & 3: Momentum bins
plot_momentum_bins(team_windows_df,
                   save_path=os.path.join(FIGURES_DIR, 'fig2_3_momentum_bins.png'))
```

```python
# Fig 4: Real vs Null Model comparison
plot_real_vs_sim(team_windows_df, sim_windows_df, bootstrap_df,
                 save_path=os.path.join(FIGURES_DIR, 'fig4_real_vs_sim.png'))
```

```python
# Fig 5 (optional): Bootstrap distributions
plot_bootstrap_intervals(bootstrap_df,
                         save_path=os.path.join(FIGURES_DIR, 'fig5_bootstrap_intervals.png'))
```
