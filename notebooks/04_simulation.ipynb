# 04 — Simulation
Run null model simulation, build sim_team_windows_df, compare real vs sim deltas, Fig 1 & 4.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import os

from config import INTERIM_DIR, BOOT_SEED
from simulate import run_simulation
from utils import compute_delta
from plots import plot_match_timeline, plot_real_vs_sim
```

```python
# Load interim data
shots_df = pd.read_csv(os.path.join(INTERIM_DIR, 'shots.csv'))
team_windows_df = pd.read_csv(os.path.join(INTERIM_DIR, 'team_windows.csv'))
print(f'shots_df: {len(shots_df)} rows')
print(f'team_windows_df: {len(team_windows_df)} rows')
```

```python
# Fig 1: Match timeline for first match
first_match_id = shots_df['match_id'].iloc[0]
print(f'Plotting timeline for match {first_match_id}')
plot_match_timeline(shots_df, first_match_id)
```

```python
# Run null model simulation
print('Running simulation...')
sim_team_windows_df = run_simulation(shots_df, seed=BOOT_SEED)
print(f'sim_team_windows_df: {len(sim_team_windows_df)} rows')
```

```python
# Compare real vs sim deltas
delta_shot_real = compute_delta(team_windows_df, 'shot_next_2')
delta_goal_real = compute_delta(team_windows_df, 'goal_next_5')
delta_shot_sim  = compute_delta(sim_team_windows_df, 'shot_next_2')
delta_goal_sim  = compute_delta(sim_team_windows_df, 'goal_next_5')

print(f'delta_shot  — Real: {delta_shot_real:.4f}  |  Null: {delta_shot_sim:.4f}  |  Gap: {delta_shot_real-delta_shot_sim:.4f}')
print(f'delta_goal  — Real: {delta_goal_real:.4f}  |  Null: {delta_goal_sim:.4f}  |  Gap: {delta_goal_real-delta_goal_sim:.4f}')
```
